```python
import requests

# URL do site que queremos raspar
url = 'http://www.escolavirtual.gov.br'

# Realizamos a requisição ao site e obtemos a resposta
resposta = requests.get(url)
```

```python
# Vamos verificar se a requisição foi bem-sucedida. Aperta o Play, Gênio.
if resposta.status_code == 200:
    print("Tudo certo!")
else:
    print("Erro. Código de Status:", resposta.status_code)
```
